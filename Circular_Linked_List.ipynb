```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

class CircularList:
  def __init__(self):
    self.head = None

  def size(self):  #Returns number of nodes. Note: Counting starts from 1.
    temp = self.head
    i = 0
    if self.head is not None:
      i = 1
    else:
      self.show()
      return i
    if self.head.next is not None:
      while not (temp.next == self.head ):
        temp = temp.next
        i += 1
    return i

  def show(self):  #Print whole Linked List along with it's index number
    if self.head is None:
      print("No item in List found! Error: {}".format(type(self.head)))
    else:
      temp = self.head
      i = 0
      print("")
      while True:  #Performs While-do functions
        if temp.next is self.head or temp.next is None:
          print("node#{} {}".format(i, temp.data))
          break
        else:
          print("node#{} {}".format(i, temp.data))
          temp = temp.next
          i += 1
      print("")
  
  def append(self, data, node = None):  # Appends at specific node/index number
    new_node = Node(data)               # By Default it appends at the last
    if self.head is None:               # i.e before head element/item in list.
      self.head = new_node
    else:
      if node is None:
        node = self.size()-1
      else:
        node-=1
      i = 0
      temp = self.head
      while True:
        if i == node:
          new_node.next = temp.next
          temp.next = new_node
          if new_node.next is None:
            new_node.next = self.head
          break
        else:
          temp = temp.next
          i += 1

  #If an argument is passed then this functions pops out the specific position
  #Otherwise it removes the most recent node (tail node) by default
  def pop(self, position = None):
    if self.size() is 0:
      return
    else:
      if position is None:
        position = self.size()-1
      if position is 0:
        self.head = self.head.next
      else:
        temp = self.head
        i = 0
        while True:
          if i+1 == position:
            temp.next = temp.next.next
            if temp.next is None:
              self.tail = temp
            break
          else:
            i += 1
            temp = temp.next

  def showOdds(self):
    if self.head is None:
      print("No item in List found! Error: {}".format(type(self.head)))
    elif self.head.next is None:
      print("No odd item in List found! Error: {}".format(type(self.head)))
    else:
      temp = self.head.next
      i = 1
      print("")
      if self.size() is 2:
        print("node#{} {}".format(i, temp.data))
      while True:  #Performs While-do functions
        if temp.next is self.head:
          break
        elif temp.next.next is self.head:
          print("node#{} {}".format(i, temp.data))
          break
        else:
          print("node#{} {}".format(i, temp.data))
          temp = temp.next.next
          i += 2
      print("")
```

```python
def menu():
  print("Please choose from following options:-")
  print("1. Insert New Node")
  print("2. Delete Node")
  print("3. Display Size of Circular Linked List")
  print("4. Display Nodes")
  print("5. Display Odd Number of Nodes")
  print("")
  print("Press any other key to exit.")
  ch = input("Enter No. of your choice: ")
  return ch
```

```python
if __name__=="__main__":
  c = CircularList()

  while True:
    choice = menu()
    if choice is '1':
      data = eval(input("Enter data for new node: "))
      position = input("Index? (Donot pass value if you want to insert at the last): ").strip()
      if position:
        c.append(data,eval(position))
      else:
        c.append(data)
    elif choice is '2':
      position = eval(input("Index? (Donot pass value if you want to delete the last node): "))
      if position:
        c.pop(eval(position))
      else:
        c.pop()
    elif choice is '3':
      print("Size of list is: {}".format(c.size()))
    elif choice is '4':
      c.show()
    elif choice is '5':
      c.showOdds()
    else:
      print("Program terminated..!")
      break
      
```
